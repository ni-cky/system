```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

fig = plt.figure(figsize=(12, 6))


for challenge in range(20, 3, -1):
    plot = [0] * 10
    for dice in range(10):
        successes = [0] * 1000
        for i in range(1000):
            throws = np.random.choice(21, size=dice + 1)
            s = 0
            for throw in throws:
                if throw >= challenge:
                    s += 1
            successes[i] = s
        mean = np.mean(successes)
        plot[dice] = mean
    plt.plot(plot, label=f'challenge {challenge}')
    
plt.xlabel('dice')
plt.ylabel('mean of succes')
plt.legend()
plt.grid()

plt.show()
```
